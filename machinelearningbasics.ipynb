SUPERVISED LEARNING- Labelled data set, understand well how input is related to output. Types- Regression- Predict continuous value output(real number)- ex- estimating house price, predicting sales, predicting market, Classification- Finite value output(countable) ex- cancerous tumor or benign tumor, hand written digit recognition, email spam or not

Unsupervised Learning- No labelled data set(no output variable). Find relation among attributes in data set. Ex- group birds of particular species in a group of mixed, differentiate music from a particular instrument among other sounds.
