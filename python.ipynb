```python
def greeting(details):
   match details:
      case [time, name]:
         return f'Good {time} {name}!'
      case [time, *names]:
         msg=''
         for name in names:
            msg+=f'Good {time} {name}!\n'
         return msg

print (greeting(["Morning", "Ravi"]))
print (greeting(["Afternoon","Guest"]))
print (greeting(["Evening", "Kajal", "Praveen", "Lata"]))
```

```python
def intr(details):
   match details:
      case [amt, duration] if amt<10000:
         return amt*10*duration/100
      case [amt, duration] if amt>=10000:
         return amt*15*duration/100
print ("Interest = ", intr([5000,5]))
print ("Interest = ", intr([15000,3]))
```

```python
def count_fruits(fruits):
  '''
  input: ['apple', 'orange', 'apple']
  output: {'apple':2, 'orange':1}
  '''
  out = {}
  for fruit in fruits:
    if fruit not in out:
      out[fruit] = 1
    else:
      out[fruit] += 1
  return out

print(count_fruits(['apple', 'orange', 'apple']))
```

```python
def is_prime(n):
  '''
  if n is a prime number, return True
  if n isn't a prime number, return False
  '''
  if(n == 1):
        return False
  else:
    for i in range(2, n//2+1):
        print(i)
        if n % i == 0:
            return False
    return True

is_prime(7)
```

```python
#program to find palindrome aba
def checkPalindrome(mystr):
    flag = True
    for i in range(0,(len(mystr)//2)+1 ):
        if mystr[i] == mystr[-i-1]:
            continue
        else:
            flag = False
            break
    if flag == True:
        print("Palindrome")
    else:
        print("Not a Palindrome")
        
checkPalindrome("absfa")
```

```python
#decorators in python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print("Something is happening before the function is called")
    return wrapper

@my_decorator
def say_hello1():
    print("Hello")
    
say_hello1()
```

```python
def hello_decorator(func):
    def inner1(*args, **kwargs):
         
        print("before Execution")
         
        # getting the returned value
        returned_value = func(*args, **kwargs)
        print("after Execution")
         
        # returning the value to the original frame
        return returned_value
         
    return inner1
 
 
# adding decorator to the function
@hello_decorator
def sum_two_numbers(a, b):
    print("Inside the function")
    return a + b
 
a, b = 1, 2
 
# getting the value through return of the function
print("Sum =", sum_two_numbers(a, b))
```
