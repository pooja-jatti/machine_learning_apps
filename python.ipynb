```python
#range function
#range(start,stop,step)
range(1,3,1)
for i in range(5):
    print(i)
print("\n")
for i in range(1,5,1):
    print(i)
print("\n")
for i in range(1,5,2):
    print(i)
print("\n")
for i in range(2,101,2): #iterate through even numbers using range
    print(i)
```

```python
#Python Dictionary are like hash table with key -value pairs, it is mutable
dict = {11:"pooja",12:"test1",13:"test2"}
print(dict)
print(dict[11])
print(dict.keys())
print(dict.values())
```

```python
#python set -collection of items but not ordered or sequenced. Set cannot have duplicate values
set1 = {1,"pooja", 2.}
print(type(set1))
print(set1)
print(set1.pop())
print(set1)
```
